```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gapminder_total.csv')
df
```

```python
# Create subset of data for one year

df_subset = df[df['Year'] == 2015]
df_subset
```

```python
df['MarkerSize'] = df['population'] / 100000
```

```python
Algeria = df_subset[df_subset['Country'] == "Algeria"]
Algeria
```

```python
China = df_subset[df_subset['Country'] == "China"]
China
```

```python
# Plot the life expectancy vs fertility rate


# sns.set(style="whitegrid") # this is the default setting
sns.set_theme(style="darkgrid")
# sns.set_theme(style="ticks")
# sns.set_theme(style="white")
# sns.set_theme(style="dark")


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fertility_rate', y='Life_Expectancy', hue='continent', data=df, palette='magma_r', 
            edgecolor='mediumpurple', alpha = 1, s=120, markers = 'x', sizes=(10,1000), size='population')


# Add title and labels
plt.title('Life Expectancy vs. Fertility Rate in 2015')
plt.xlabel('Fertility Rate')
plt.ylabel('Life Expectancy')

# Show the plot
plt.legend(frameon=False, bbox_to_anchor=(1,1),);
plt.show()


```
