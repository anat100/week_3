```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Filter data for years between 1960 and 2015:

df = pd.read_csv('gapminder_total.csv')
df
```

```python
df_subset = df[df["year"] == 1950]
df_subset
```

```python
for year in range(1960, 2016):
    plot = df[df["year"]==year]
    plt.figure(figsize= (10,6))
    plt.axis([0, 85, 0, 9.5])
    sns.set_style("darkgrid")
    sns.scatterplot(x="Life_Expectancy", y="fertility", data= plot, hue= "continent", alpha=0.9, size= "population").set(title= "Life Expectancy vs Fertility Rate")
    plt.xticks(rotation= 45)
    plt.legend(loc= "lower left", frameon= False)
    plt.savefig(f'../data/plot{year}.png') 
    plt.close()
    
```

```python
import imageio.v2 as imageio

images = []

for year in range(1960, 2016):
    filename = f'../data/plot{year}.png'
    images.append(imageio.imread(filename))

imageio.mimsave('../data/life_expectancy_and_fertility_new.gif', images, duration=200)
```
