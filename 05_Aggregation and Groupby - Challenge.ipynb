```python
import pandas as pd
import numpy as np
```

```python
# 1. Read in data:

df = pd.read_csv('gapminder_total.csv')
df
```

```python
# 2. What is the median population in the data set?

median_life_exp = df['Life_Expectancy'].median()

median_life_exp
```

```python
# 3. How often does each continent appear in the data set?

continent_counts = df['continent'].value_counts()

continent_counts
```

```python
# 4. Which continent has the lowest average fertility rate overall?

continent_avg_fertility = df.groupby('continent')['Fertility_rate'].mean()
continent_avg_fertility
```

```python
lowest_avg_fertility_continent = continent_avg_fertility.idxmin()
lowest_avg_fertility = continent_avg_fertility.min()

print("Continent with lowest average fertility rate:", lowest_avg_fertility_continent)
print("Lowest average fertility rate:", lowest_avg_fertility)
```

```python
# 5. What was the average life expectancy in Europe in 2015? 
# Hint: first filter for 2015 then apply groupby.

# Filter for the year 2015

df_2015_x = df[df['Year'] == 2015]
df_2015_x

```

```python
# Filter for the 'Europe' continent in 2015

df[df['Year']==2015].groupby('continent')['Life_Expectancy'].agg('mean')
```
